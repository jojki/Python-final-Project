```python
# TASK 3: Calculate the distances between a launch site to its proximities

import pandas as pd
import numpy as np

# Define a function to calculate the Haversine distance
def haversine(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    # Radius of Earth in kilometers (mean radius)
    r = 6371.0
    return c * r

# Example DataFrame of points of interest with their coordinates
poi_df = pd.DataFrame({
    'Name': ['City A', 'Airport B', 'Port C'],
    'Lat': [28.5383, 29.9844, 27.9506],
    'Long': [-81.3792, -95.3414, -82.4572]
})

# Calculate distances from each launch site to each point of interest
distances = []
for index, launch in spacex_df.iterrows():
    for _, poi in poi_df.iterrows():
        distance = haversine(launch['Lat'], launch['Long'], poi['Lat'], poi['Long'])
        distances.append({
            'Launch Site': launch['Launch Site'],
            'Point of Interest': poi['Name'],
            'Distance (km)': distance
        })

# Create a DataFrame with the calculated distances
distances_df = pd.DataFrame(distances)

```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Launch site coordinates
launch_site_coords = (34.632834, -120.610746)

# Coordinates for points of interest
poi_coords = {
    'Closest City': (34.639150, -120.457940),
    'Closest Railway': (34.635000, -120.624000),
    'Closest Highway': (34.637800, -120.620500)
}


import numpy as np

def haversine(coord1, coord2):
    """
    Calculate the great-circle distance between two points on the Earth's surface.
    Parameters:
    coord1, coord2: Tuples containing latitude and longitude of the two points in decimal degrees.
    Returns:
    Distance between the two points in kilometers.
    """
    lat1, lon1 = coord1
    lat2, lon2 = coord2
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    # Radius of Earth in kilometers (mean radius)
    r = 6371.0
    return c * r



import folium
from folium.features import DivIcon

# Initialize the map centered at the launch site
site_map = folium.Map(location=launch_site_coords, zoom_start=14)

# Add a marker for the launch site
folium.Marker(
    location=launch_site_coords,
    popup='Launch Site',
    icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
).add_to(site_map)

# Add markers for each point of interest and draw lines
for poi, coords in poi_coords.items():
    # Calculate the distance to the launch site
    distance = haversine(launch_site_coords, coords)
    
    # Add a marker for the point of interest
    folium.Marker(
        location=coords,
        popup=f'{poi}<br>Distance: {distance:.2f} km',
        icon=folium.Icon(color='green')
    ).add_to(site_map)
    
    # Add a line between the launch site and the point of interest
    folium.PolyLine(
        locations=[launch_site_coords, coords],
        color='red',
        weight=2,
        dash_array='5,10'
    ).add_to(site_map)
    
    # Add a marker to display the distance at the midpoint
    midpoint = [(launch_site_coords[0] + coords[0]) / 2, (launch_site_coords[1] + coords[1]) / 2]
    folium.Marker(
        location=midpoint,
        icon=DivIcon(
            icon_size=(150, 36),
            icon_anchor=(75, 18),
            html=f'<div style="font-size: 12pt; color: black; background-color: white; padding: 5px; border-radius: 5px;">{distance:.2f} km</div>'
        )
    ).add_to(site_map)

# Display the map
site_map
```
